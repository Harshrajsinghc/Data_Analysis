```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('zomato_dataset.csv')
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.drop(['Best Seller'],inplace= True, axis=1)
```

```python
mean = np.mean(df['Dining Rating'])
mean
```

```python
df['Dining Rating']= df['Dining Rating'].replace(np.nan,3.8)
```

```python
df.isnull().sum()
```

```python
print(np.mean(df['Delivery Rating']))
```

```python
df['Delivery Rating']= df['Delivery Rating'].replace(np.nan,3.9)
```

```python
df.describe()
```

```python
df.duplicated().sum()
```

```python
df=df.drop_duplicates()
```

```python
df.duplicated().sum()
```

```python
df.to_csv('Zomato_data.csv')
```

### Which metropolitan area has the highest average delivery rating

```python
gb = df.groupby('City')
```

```python
gb.mean()
```

From the above table we can see City Pune has highest average delivery rating.

### What is the total number of dining votes received by all restaurants in each city

```python
df.groupby(['City','Restaurant Name'])['Dining Votes'].sum()
```

Above table has total number of dining votes of each restaurant of the city

```python
gb.sum()
```

Above table has total number of dining votes for all restaurants in the city.

###  Which restaurant has the highest average dining rating in each city
